```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crude_oil_data = pd.read_csv('crude_oil_data_processed.csv')
crude_oil_data.head(5)
```

```python
crude_oil_data['Date'] = pd.to_datetime(crude_oil_data['Date'])
crude_oil_data['Date'].head()
```

```python
crude_oil_data.columns
```

```python
crude_oil_data.plot(x='Date', y='U.S. Crude Oil ', figsize=(12,8), color='brown')
plt.ylabel('Production')
plt.title('U.S. Crude Oil Production')
```

```python
crude_oil_data.boxplot('U.S. Crude Oil ', figsize=(12,8)) #best viz to use to view summary descriptive statistics for dataset is boxplot
plt.ylabel('Production')
plt.title('U.S. Crude Oil Production')
```

```python
crude_oil_data.boxplot('California', figsize=(12,8)) #other syntax: crude_oil_data[['California']].boxplot(figsize=(12,8))
plt.ylabel('Production in California')
plt.title('California Crude Oil Production')
```

```python
crude_oil_data[['Alaska', 'California']].boxplot(figsize=(12,8))
plt.ylabel('Production in Alaska and California')
plt.title('Alaska and California Crude Oil Production')
```

```python
crude_oil_data.boxplot('U.S. Crude Oil ', by=['Year'], figsize=(12,8))
plt.ylabel('Production')
```

```python
year_data = crude_oil_data.groupby('Year', as_index = False).sum() #as_index=False will ensure col 'Year' does not become the index
```

```python
colors=['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']
year_data.plot.bar(x='Year', y='U.S. Crude Oil ', figsize=(12,8), color=colors)
```

```python
crude_oil_data
```

```python
mean_prod_data = crude_oil_data.mean()[1:-3]
mean_prod_data
```

```python
mean_prod_data = mean_prod_data.sort_values(ascending=False) #sort_values syntax has 'ascending'
mean_prod_data
```

```python
mean_prod_df = pd.DataFrame(mean_prod_data).reset_index()
mean_prod_df.columns = ['State', 'Production']
mean_prod_df.head(10)
```

```python
plt.figure(figsize=(12,8))
plt.bar(mean_prod_df['State'], mean_prod_df['Production'], width = 0.85)
plt.title('US Oil Mean-Production June 2008 to June 2018')
plt.xticks(rotation=90)
#plt.yticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Production')
```

```python
mean_prod_df['Production'].plot.kde(figsize=(12,8)) #KDE (Kernel Density Estimation)
plt.xlabel('Oil Production')
plt.title('US Oil Production June 2008 to June 2018')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(crude_oil_data['Texas'], crude_oil_data['U.S. Crude Oil '], c='g') #'g' is for Green
plt.xlabel('Texas Production')
plt.ylabel('US Production')
plt.show()
```
